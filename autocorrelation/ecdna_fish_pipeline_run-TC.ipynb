```python
import sys
import subprocess
from pathlib import Path

# =========================
# 0) 用户可改参数区
# =========================
GROUP = "MYC_PVT1"

INPUT_ROOT = Path(r"D:\02.project\18.ecDNA\16.实验\04.染色结果\WBSY\MYC_PVT1")
OUT_ROOT   = Path(r"D:\02.project\18.ecDNA\02.code\v0.1.1\autocorrelation\output_dc\MYC_PVT1")

# 脚本位置：改为你本机实际路径
PIPELINE_PY = Path(r"D:\02.project\18.ecDNA\02.code\v0.1.1\autocorrelation\ecdna_fish_pipeline_v4.py")
AUTO_PY     = Path(r"D:\02.project\18.ecDNA\02.code\v0.1.1\autocorrelation\ecdna_autocorrelation_v4.py")

RUN = True  # 先 False 打印；确认无误后改 True 执行

# 允许的图片扩展名（按需增减）
EXTS = [".tif", ".tiff", ".bmp", ".jpg", ".jpeg", ".png"]

# 每个ROI所需5张图的“基名”到参数映射
REQUIRED = {
    "DAPI": "dapi",
    "R": "red",
    "G": "green",
    "A": "cyan",
    "RGA": "merged",
}

# =========================
# 1) 固定参数（与示例保持一致）
# =========================
pipeline_fixed = [
    "--nuclei_method", "stardist",
    "--min_area_px", "800",
    "--cellpose_diameter", "80",
    "--stardist_prob_thresh", "0.35",
    "--stardist_nms_thresh", "0.5",
    "--stardist_sigma", "1.5",
    "--stardist_merge_quantile", "0.95",
    "--rescue_max_region_area_px", "120000",
    "--rescue_h_maxima_h", "0.5",
    "--fish_smooth_sigma", "1.2",
    "--fish_bg_sigma", "6",
    "--spot_radius_px", "3",
    "--use_auto_threshold",
    "--auto_thr_factor", "0.8",
    "--enable_dense",
    "--dense_z", "20",
    "--max_extra_spots_per_component", "2",
    "--centromere_max_spots", "4",
    "--centromere_disable_dense",
    "--coloc_dist_px", "10",
    "--qc_nuclei", "10",
    "--qc_seed", "0",
    "--dapi_rgb", "auto",
    "--red_rgb", "r",
    "--green_rgb", "g",
    "--cyan_rgb", "max",
    "--tif_first_page",
]

autocorr_fixed = [
    "--tif_first_page",
    "--dapi_rgb", "b", "--dapi_z_project", "max",
    "--rmax", "100", "--n_random", "5", "--n_dapi", "3", "--seed", "0",
    "--spot_kernel", "gaussian", "--spot_sigma_px", "5",
    "--plot", "--plot_rmax", "50", "--plot_min_spots", "2", "--plot_bin_px", "1",
]

# =========================
# 2) 工具函数
# =========================
def show_cmd(cmd):
    def q(x):
        if x.startswith("--"):
            return x
        return f'"{x}"' if " " in x else x
    print(" ".join(q(str(x)) for x in cmd))

def run_cmd(cmd, tag=""):
    show_cmd(cmd)
    if not RUN:
        return True

    p = subprocess.run(cmd, text=True, capture_output=True)
    if p.returncode == 0:
        # 如果你不想刷屏，可以注释掉下面两行
        if p.stdout:
            print(p.stdout)
        if p.stderr:
            print(p.stderr)
        return True

    # 失败：打印完整输出（或至少末尾）
    print(f"\n[ERROR] {tag} returncode={p.returncode}")
    if p.stdout:
        print("\n--- STDOUT ---\n", p.stdout[-8000:])  # 打印末尾，避免太长
    if p.stderr:
        print("\n--- STDERR ---\n", p.stderr[-8000:])
    return False
    
def find_ext(raw_dir: Path):
    """在 raw_dir 中找一个可用扩展名，使 REQUIRED 的所有基名文件都存在"""
    for ext in EXTS:
        ok = True
        for base in REQUIRED:
            if not (raw_dir / f"{base}{ext}").exists():
                ok = False
                break
        if ok:
            return ext
    return None

# =========================
# 3) 扫描样本/ROI并组装任务
# =========================
tasks = []
if not INPUT_ROOT.exists():
    raise FileNotFoundError(f"INPUT_ROOT not found: {INPUT_ROOT}")

for sample_dir in sorted([p for p in INPUT_ROOT.iterdir() if p.is_dir()]):
    sample_name = sample_dir.name  # e.g., BRCA_P1
    for roi_dir in sorted([p for p in sample_dir.iterdir() if p.is_dir()]):
        roi = roi_dir.name          # e.g., 1
        raw_dir = roi_dir / "原始图片"
        if not raw_dir.exists():
            continue

        ext = find_ext(raw_dir)
        if ext is None:
            # 给出缺哪些（按每个ext都可能缺，不深究；直接列出存在的文件帮助定位）
            existing = sorted([x.name for x in raw_dir.iterdir() if x.is_file()])
            print(f"[SKIP] {sample_name}/{roi}: 未找到完整5张图（DAPI/R/G/A/RGA 同一后缀）。现有文件: {existing[:10]}{' ...' if len(existing)>10 else ''}")
            continue

        files = {arg: str(raw_dir / f"{base}{ext}") for base, arg in REQUIRED.items()}
        sample_id = f"{sample_name}_ROI{roi}"

        tasks.append({
            "sample_id": sample_id,
            "files": files,
            "dapi": files["dapi"],
        })

print(f"[FOUND] {len(tasks)} ROI(s) ready.")

# =========================
# 4) 批量运行两步流程
# =========================
OUT_ROOT.mkdir(parents=True, exist_ok=True)

for t in tasks:
    sample_id = t["sample_id"]
    files = t["files"]

    print(f"\n[STEP1] {sample_id}")
    cmd1 = [
        sys.executable, PIPELINE_PY,
        "--outdir", OUT_ROOT,
        "--group", GROUP,
        "--sample_id", sample_id,
        "--dapi", files["dapi"],
        "--red", files["red"],
        "--green", files["green"],
        "--cyan", files["cyan"],
        "--merged", files["merged"],
        *pipeline_fixed,
    ]
    #run_cmd(cmd1)
    ok1 = run_cmd(cmd1, tag=f"STEP1 {sample_id}")
    if not ok1:
        continue  # 该ROI第一步失败，跳过第二步，继续下一个ROI

    print(f"[STEP2] {sample_id}")
    cmd2 = [
        sys.executable, AUTO_PY,
        "--outdir", OUT_ROOT / sample_id,
        "--dapi", t["dapi"],
        *autocorr_fixed,
    ]
    #run_cmd(cmd2)
    ok2 = run_cmd(cmd2, tag=f"STEP2 {sample_id}")
    if not ok2:
        continue

print("\n[DONE] dry-run finished." if not RUN else "\n[DONE] all jobs finished.")
```

```python
### "--nuclei_method", "watershed" ws

import sys
import subprocess
from pathlib import Path

# =========================
# 0) 用户可改参数区
# =========================
GROUP = "MYC_PVT1"

INPUT_ROOT = Path(r"D:\02.project\18.ecDNA\16.实验\04.染色结果\WBSY\MYC_PVT1")
OUT_ROOT   = Path(r"D:\02.project\18.ecDNA\02.code\v0.1.1\autocorrelation\output_ws\MYC_PVT1")

# 脚本位置：改为你本机实际路径
PIPELINE_PY = Path(r"D:\02.project\18.ecDNA\02.code\v0.1.1\autocorrelation\ecdna_fish_pipeline_v4.py")
AUTO_PY     = Path(r"D:\02.project\18.ecDNA\02.code\v0.1.1\autocorrelation\ecdna_autocorrelation_v4.py")

RUN = True  # 先 False 打印；确认无误后改 True 执行

# 允许的图片扩展名（按需增减）
EXTS = [".tif", ".tiff", ".bmp", ".jpg", ".jpeg", ".png"]

# 每个ROI所需5张图的“基名”到参数映射
REQUIRED = {
    "DAPI": "dapi",
    "R": "red",
    "G": "green",
    "A": "cyan",
    "RGA": "merged",
}

# =========================
# 1) 固定参数（与示例保持一致）
# =========================
pipeline_fixed = [
    "--nuclei_method", "watershed",
    "--min_area_px", "800",
    "--cellpose_diameter", "80",
    "--stardist_prob_thresh", "0.35",
    "--stardist_nms_thresh", "0.5",
    "--stardist_sigma", "1.5",
    "--stardist_merge_quantile", "0.95",
    "--rescue_max_region_area_px", "120000",
    "--rescue_h_maxima_h", "0.5",
    "--fish_smooth_sigma", "1.2",
    "--fish_bg_sigma", "6",
    "--spot_radius_px", "3",
    "--use_auto_threshold",
    "--auto_thr_factor", "0.8",
    "--enable_dense",
    "--dense_z", "20",
    "--max_extra_spots_per_component", "2",
    "--centromere_max_spots", "4",
    "--centromere_disable_dense",
    "--coloc_dist_px", "10",
    "--qc_nuclei", "10",
    "--qc_seed", "0",
    "--dapi_rgb", "auto",
    "--red_rgb", "r",
    "--green_rgb", "g",
    "--cyan_rgb", "max",
    "--tif_first_page",
]

autocorr_fixed = [
    "--tif_first_page",
    "--dapi_rgb", "b", "--dapi_z_project", "max",
    "--rmax", "100", "--n_random", "5", "--n_dapi", "3", "--seed", "0",
    "--spot_kernel", "gaussian", "--spot_sigma_px", "5",
    "--plot", "--plot_rmax", "50", "--plot_min_spots", "2", "--plot_bin_px", "1",
]

# =========================
# 2) 工具函数
# =========================
def show_cmd(cmd):
    def q(x):
        if x.startswith("--"):
            return x
        return f'"{x}"' if " " in x else x
    print(" ".join(q(str(x)) for x in cmd))

def run_cmd(cmd, tag=""):
    show_cmd(cmd)
    if not RUN:
        return True

    p = subprocess.run(cmd, text=True, capture_output=True)
    if p.returncode == 0:
        # 如果你不想刷屏，可以注释掉下面两行
        if p.stdout:
            print(p.stdout)
        if p.stderr:
            print(p.stderr)
        return True

    # 失败：打印完整输出（或至少末尾）
    print(f"\n[ERROR] {tag} returncode={p.returncode}")
    if p.stdout:
        print("\n--- STDOUT ---\n", p.stdout[-8000:])  # 打印末尾，避免太长
    if p.stderr:
        print("\n--- STDERR ---\n", p.stderr[-8000:])
    return False
    
def find_ext(raw_dir: Path):
    """在 raw_dir 中找一个可用扩展名，使 REQUIRED 的所有基名文件都存在"""
    for ext in EXTS:
        ok = True
        for base in REQUIRED:
            if not (raw_dir / f"{base}{ext}").exists():
                ok = False
                break
        if ok:
            return ext
    return None

# =========================
# 3) 扫描样本/ROI并组装任务
# =========================
tasks = []
if not INPUT_ROOT.exists():
    raise FileNotFoundError(f"INPUT_ROOT not found: {INPUT_ROOT}")

for sample_dir in sorted([p for p in INPUT_ROOT.iterdir() if p.is_dir()]):
    sample_name = sample_dir.name  # e.g., BRCA_P1
    for roi_dir in sorted([p for p in sample_dir.iterdir() if p.is_dir()]):
        roi = roi_dir.name          # e.g., 1
        raw_dir = roi_dir / "原始图片"
        if not raw_dir.exists():
            continue

        ext = find_ext(raw_dir)
        if ext is None:
            # 给出缺哪些（按每个ext都可能缺，不深究；直接列出存在的文件帮助定位）
            existing = sorted([x.name for x in raw_dir.iterdir() if x.is_file()])
            print(f"[SKIP] {sample_name}/{roi}: 未找到完整5张图（DAPI/R/G/A/RGA 同一后缀）。现有文件: {existing[:10]}{' ...' if len(existing)>10 else ''}")
            continue

        files = {arg: str(raw_dir / f"{base}{ext}") for base, arg in REQUIRED.items()}
        sample_id = f"{sample_name}_ROI{roi}"

        tasks.append({
            "sample_id": sample_id,
            "files": files,
            "dapi": files["dapi"],
        })

print(f"[FOUND] {len(tasks)} ROI(s) ready.")

# =========================
# 4) 批量运行两步流程
# =========================
OUT_ROOT.mkdir(parents=True, exist_ok=True)

for t in tasks:
    sample_id = t["sample_id"]
    files = t["files"]

    print(f"\n[STEP1] {sample_id}")
    cmd1 = [
        sys.executable, PIPELINE_PY,
        "--outdir", OUT_ROOT,
        "--group", GROUP,
        "--sample_id", sample_id,
        "--dapi", files["dapi"],
        "--red", files["red"],
        "--green", files["green"],
        "--cyan", files["cyan"],
        "--merged", files["merged"],
        *pipeline_fixed,
    ]
    #run_cmd(cmd1)
    ok1 = run_cmd(cmd1, tag=f"STEP1 {sample_id}")
    if not ok1:
        continue  # 该ROI第一步失败，跳过第二步，继续下一个ROI

    print(f"[STEP2] {sample_id}")
    cmd2 = [
        sys.executable, AUTO_PY,
        "--outdir", OUT_ROOT / sample_id,
        "--dapi", t["dapi"],
        *autocorr_fixed,
    ]
    #run_cmd(cmd2)
    ok2 = run_cmd(cmd2, tag=f"STEP2 {sample_id}")
    if not ok2:
        continue

print("\n[DONE] dry-run finished." if not RUN else "\n[DONE] all jobs finished.")
```
