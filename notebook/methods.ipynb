```python
import os
import gzip

def compress_files(folder_path):
    # 获取文件夹中的所有文件
    files = os.listdir(folder_path)

    # 遍历文件
    for file_name in files:
        # 检查文件是否以".gz"为后缀
        if not file_name.endswith(".gz"):
            # 构建文件的完整路径
            file_path = os.path.join(folder_path, file_name)

            # 构建压缩后的文件路径
            gz_file_path = file_path + ".gz"

            # 打开原文件进行读取，并使用 gzip 进行压缩写入
            with open(file_path, 'rb') as f_in, gzip.open(gz_file_path, 'wb') as f_out:
                f_out.writelines(f_in)

            print(f"文件 '{file_name}' 已压缩为 '{gz_file_path}'")
            pass
        pass
    pass
```

```python
def compare_files(file1, file2):
    with open(file1, 'r') as a, open(file2, 'r') as b:
        set_a = {tuple(line.strip().split('/')[-2:]) for line in a}
        set_b = {tuple(line.strip().split('/')[-2:]) for line in b}

    common = set_a & set_b
    diff = (set_a - set_b).union(set_b - set_a)

    with open('common.txt', 'w') as f:
        for item in common:
            f.write('/'.join(item) + '\n')

    with open('diff.txt', 'w') as f:
        for item in diff:
            f.write('/'.join(item) + '\n')

compare_files('a.txt', 'c1.txt')
```

```python
def split_file(filename, lines_per_file):
    with open(filename, 'r') as f:
        lines = f.readlines()
        
    for i in range(0, len(lines), lines_per_file):
        small_filename = f"{filename}_{i//lines_per_file}.txt"
        with open(small_filename, 'w') as f:
            f.writelines(lines[i:i+lines_per_file])

split_file('D:\\02.project\\29.CRC\\数据清单\\diff-all.txt', 100)
```

```python
#windows python 遍历某路径下的子文件夹，子文件夹中没有文件的路径筛选出来，组成一个list
import os

def find_empty_subfolders(base_path):
    empty_subfolders = []

    for root, dirs, files in os.walk(base_path):
        if not files:  # 如果文件列表为空，说明该子文件夹没有文件
            empty_subfolders.append(root)

    return empty_subfolders

# 替换为你要遍历的目标路径
target_path = r'G:/BGItrans/'

empty_subfolders_list = find_empty_subfolders(target_path)

dfempty = pd.DataFrame({'file_path': empty_subfolders_list})
dfempty = dfempty.map(lambda x: x.replace('\\', '/'))

print(dfempty["file_path"][2])
```
